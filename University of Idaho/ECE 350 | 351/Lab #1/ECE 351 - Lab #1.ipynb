## Part 2b - Using Python 3.x | Plotting Functions
The general idea to plotting functions in Python is to use the `matplotlib` package, and create a `pyplot` object. This object can have one or more plots (individual graphs) and each plot can have many different traces. These objects can be incredibly complex, but for now we were just introducted to the basics of plotting. This is shown in the provided example code with a single figure of a fixed size with three subplots. These all have the same axis, but different scales and plots each time.

I've attached the provided code for plotting functions. Each of the described components are listed.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# Define Variables
steps = 0.1
x = np.arange(-2, 2 + steps, steps)
y1 = x + 2
y2 = x**2

# Code for Plots
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(x, y1)
plt.title("Sample Plots for Lab #1")

plt.ylabel('Subplot 1')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(x, y2)
plt.ylabel("Subplot 2")
plt.grid(which="both")

plt.subplot(3, 1, 3)
plt.plot(x, y1, "--r", label="y1")
plt.plot(x, y2, "o", label="y2")
plt.axis([-2.5, 2.5, -0.5, 4.5])
plt.grid(True)
plt.legend(loc="lower right")
plt.xlabel("x")

plt.ylabel("Subplot 3")

plt.show()
```
