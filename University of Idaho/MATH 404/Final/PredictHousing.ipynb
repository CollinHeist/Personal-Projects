##House price predictions from the Ames housing dataset.

Library imports and data input...

Read in data as dataframe
See link for info on dataset. https://www.kaggle.com/prevek18/ames-housing-dataset Gives a breakdown on the contents of each column.

```
import pandas as pd
import numpy as np
from sklearn import preprocessing

import keras
import keras.backend as kb
import tensorflow as tf


from google.colab import drive
drive.mount('/content/drive')

Ames = pd.read_csv('/content/drive/My Drive/404 Project/AmesHousing_Reduced.csv')
Labels = Ames.pop('SalePrice')
Ames = Ames.drop('Bsmt Full Bath',axis=1)
Ames = Ames.drop('Bsmt Half Bath',axis=1)

# begin to put columns into categories
# to start I used 10 categories

# LotBins is a tuple interpretation of intervals
LotBins = pd.IntervalIndex.from_tuples([(1000,6500), (6501,8500), (8501,10000), (10001,11500), (11501,215500)])
Ames['Lot Area'] = pd.cut(Ames['Lot Area'], LotBins, labels=False)


Ames['Year Built'] = pd.cut(Ames['Year Built'], 10, labels=False)
Ames['Year Remod/Add'] = pd.cut(Ames['Year Remod/Add'], 10, labels=False)
Ames['Total Bsmt SF'] = pd.cut(Ames['Total Bsmt SF'], 10, labels=False)
Ames['1st Flr SF'] = pd.cut(Ames['1st Flr SF'], 10, labels=False)
Ames['2nd Flr SF'] = pd.cut(Ames['2nd Flr SF'],  10, labels=False)
Ames['Gr Liv Area'] = pd.cut(Ames['Gr Liv Area'], 10, labels=False)
Ames['Garage Area'] = pd.cut(Ames['Garage Area'], 10, labels=False)

Labels = pd.cut(Labels,50,labels=False)
```

# Create a one-hot encoded dataset.


```
Ames1Hot = Ames

cols = ['Yr Sold','Garage Cars','Full Bath','Half Bath','Bedroom AbvGr','TotRms AbvGrd','Lot Area','Year Built','Year Remod/Add','Total Bsmt SF','1st Flr SF','2nd Flr SF','Gr Liv Area','Garage Area']
Ames1Hot = pd.get_dummies(Ames1Hot,columns = cols)


Labels1Hot = Labels
Labels1Hot = pd.get_dummies(Labels1Hot)
```

#Split the validation set off

```
X_validation = Ames1Hot.sample(frac=0.05, random_state=10)
Y_validation = Labels1Hot.sample(frac = 0.05, random_state=10)
X = Ames1Hot.to_numpy()
Y = Labels1Hot.to_numpy()
```

#Model Definition and training

```
# Second attempt at a regression model. I think we have too many variables, or I have done a poor job at prepping the data.

model = keras.Sequential([
    keras.layers.Dense(32, activation='relu', input_dim=X.shape[1]),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(39, activation='softmax')
  ])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, epochs=1000, verbose=1, validation_data=(X_validation, Y_validation))
```
